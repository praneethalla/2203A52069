```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('/Salary_Data.csv')
print(df)
```

```python
X= df['Salary'].values
Y= df['YearsExperience'].values
```

```python
mean_x =np.mean(X)
mean_y =np.mean(Y)
n=len(X)
b1 =0
b2=0
for i in range(n):
  b1 += (X[i] - mean_x) * (Y[i] - mean_y)
  b2 += (X[i] - mean_x) ** 2
m1 = b1 / b2
m0 = mean_y - (m1 * mean_x)
print(m1,m0)
```

```python
max_x =np.max(X) + 100
min_x = np.min(Y) - 100
x= np.linspace(min_x,max_x,1000)
y =m0 +m1* X
plt.plot(X,Y, color='#58b970', label='regression line')
plt.scatter( X,Y,color='#ef5423', label='scatter plot')
plt.xlabel('Salary')
plt.ylabel('YearsExperience')
plt.legend()
plt.show()
```
